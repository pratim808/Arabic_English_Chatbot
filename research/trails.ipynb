```python
from transformers import pipeline
import time
import os
from langchain.chains import create_retrieval_chain
from transformers import BartForConditionalGeneration, BartTokenizer
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from dotenv import load_dotenv

# Load the FAISS vector database
# embeddings = HuggingFaceEmbeddings(model_name="omarelshehy/arabic-english-sts-matryoshka-v2.0")
# db = FAISS.load_local(DB_FAISS_PATH, embeddings, allow_dangerous_deserialization=True)
# load_dotenv()

# Load the BART model and tokenizer
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large")

# Create HuggingFacePipeline
bart_pipeline = pipeline("text2text-generation",
                         model=model, tokenizer=tokenizer)

# Convert HuggingFace pipeline to LangChain LLM
llm = HuggingFacePipeline(pipeline=bart_pipeline)

# Define the prompt template
prompt_template = """Use the following pieces of information to answer the user's question.
If you don't know the answer, just say that you don't know, don't try to make up an answer.
Context: {context}
Question: {input}
Only return the helpful answer below and nothing else.
Helpful answer:"""

prompt = PromptTemplate(
    input_variables=["context", "input"], template=prompt_template)

# Get user input
prompt1 = input("Enter Your Question From Documents: ")

# Use FAISS as a retriever
vectors = db
print("Vector Store DB Is Ready")

if prompt1:
    # Create the LLM chain
    llm_chain = LLMChain(llm=llm, prompt=prompt)

    # Create the retriever and retrieval chain
    retriever = vectors.as_retriever()
    retrieval_chain = create_retrieval_chain(retriever, llm_chain)

    # Start the retrieval and response process
    start = time.process_time()
    response = retrieval_chain.invoke({'input': prompt1})
    print("Response time:", time.process_time() - start)
    print(response['answer'])

    # Print similar documents
    print("\nDocument Similarity Search:")
    for i, doc in enumerate(response["context"]):
        print(doc.page_content)
        print("--------------------------------")
```
